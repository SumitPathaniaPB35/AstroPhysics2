## 1D model Fitting
linear fit 

m*X+c

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy 
from astropy.modeling import models,fitting
import random
```

```python
x=np.linspace(0,400,10)
y=x+np.random.randint(0,50,10)
plt.scatter(x,y,s=120)
```

```python
plt.scatter(x,y,s=120)
fit=fitting.LinearLSQFitter()
line_fit=models.Linear1D()
fitted_line=fit(line_fit,x,y)
plt.plot(x,fitted_line(x),c='red',lw=5)
plt.title('Astro 1D model fit')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
print(fitted_line)
```
