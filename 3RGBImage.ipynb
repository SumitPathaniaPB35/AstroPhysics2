## RGB Image from Fits FIle

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# For RGB

from astropy.visualization import make_lupton_rgb
```

```python
r=fits.open('frame-r-005071-3-0347.fits')
i=fits.open('frame-i-005071-3-0347.fits')
g=fits.open('frame-g-005071-3-0347.fits')
```

```python
r_data=r[0].data
i_data=i[0].data
g_data=g[0].data

```

```python
plt.imshow(np.log(r_data),origin='lower')
plt.show()
```

```python
plt.imshow(np.log(i_data),origin='lower')
plt.show()
```

```python
plt.imshow(np.log(g_data),origin='lower')
plt.show()
```

```python
rgb=make_lupton_rgb(r_data,i_data,g_data,Q=10,stretch=0.5) # Q and Streatch are there to have a more clearer image
plt.imshow(rgb,origin='lower')
plt.show()
```
